```python
import math
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
rowa = pd.read_csv('rowa_results.csv')
```

```python
clients_number = [1, 2, 3, 5]
put_interval = ['2000ms', '1000ms', '500ms', '200ms', '100ms']
for i in clients_number:
        fig, ax = plt.subplots()
        if(i == 1):
             ax.set_title('Config ROWA, with one client' )
             rowa[(rowa['clientsNumber'] == i)].to_csv('config_rowa_%i_client.csv' % i)
        else:
            ax.set_title('Config ROWA, with %i clients' % i)
            rowa[(rowa['clientsNumber'] == i)].to_csv('config_rowa_%i_clients.csv' % i)
        ax.set_xlabel('keyspaceSize')
        ax.set_ylabel('avg')
        for j in put_interval:
            temp_ks = pd.Series(rowa[(rowa['clientsNumber'] == i) & (rowa['putInterval'] == '%s' % j)]['keyspaceSize'])
            temp_avg = pd.Series(rowa[(rowa['clientsNumber'] == i) & (rowa['putInterval'] == '%s' % j)]['avg'])
            ax.plot(temp_ks, temp_avg, label = '%s' % j, linewidth=0.75)
            ax.legend()
            plt.savefig('config_rowa_%i_client.png' % i)
```
